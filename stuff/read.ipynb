```python
import pandas as pd
import os

# Define file paths
file_a = r'c:\Tejas\Wind\wind_turbine_snippet_A.csv'
file_b = r'c:\Tejas\Wind\wind_turbine_snippet_B.csv'

def read_and_display(file_path, label):
    print(f"\n--- {label} (First 5 rows) ---")
    if os.path.exists(file_path):
        try:
            df = pd.read_csv(file_path)
            print(df.head())
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
    else:
        print(f"File not found: {file_path}")

if __name__ == "__main__":
    read_and_display(file_a, "Dataset A")
    read_and_display(file_b, "Dataset B")
```
